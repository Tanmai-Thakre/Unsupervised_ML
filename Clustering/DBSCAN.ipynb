### For Iris flower dataset

```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
# Scaling 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
sns.scatterplot(x=X[:,0] , y=X[:,2])
```

```python
# DBSCAN 
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps = 0.8,
    min_samples=5
)
label = dbscan.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0] , y=X[:,2] , c=label)
```

### For non-linear data

```python
from sklearn.datasets import make_moons

X , y = make_moons(
    n_samples = 300,
    noise=0.05,
    random_state=42
)
```

```python
#Scaling 
X_scaled = scaler.fit_transform(X)
```

```python
sns.scatterplot(x=X[:,0] , y=X[:,1] )
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
```

```python
# Implementing k-means clustering 
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=2,
    random_state=42
)
k_label = kmeans.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X[:,0] , y=X[:,1] ,c=k_label)
```

```python
# Implementing for DBSCAN
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)
db_label = dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X[:,0] , y=X[:,1] ,c=db_label)
```
